```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('corporate_stress_dataset.csv')
df.head()
```

```python
marital_status_counts = df['Marital_Status'].value_counts()
grouped_marital_status = df.groupby(['Marital_Status'])['Stress_Level'].mean()
x = grouped_marital_status.index
y = grouped_marital_status.values
plt.bar(x,y)
plt.title('Marital Status Stress Distribution')
plt.show()
```

```python
gender_counts = df.loc[df['Gender']=="Non-Binary"]
gender_counts
Grouped_gender_genderbias = gender_counts.groupby(['Gender', 'Gender_Bias_Experienced']).size()
Grouped_gender_genderbias
plt.pie(Grouped_gender_genderbias, labels=Grouped_gender_genderbias.index, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightcoral', 'lightgreen'])
plt.title('Gender Bias Experience Distribution')
plt.show()
```

```python
bins = [-1, 3, 7, 10]  
labels = ["Low", "Medium", "High"]


df["Stress Category"] = pd.cut(df["Stress_Level"], bins=bins, labels=labels)


result = df.groupby(["Marital_Status", "Stress Category"]).size().reset_index(name="Count")
result
```
